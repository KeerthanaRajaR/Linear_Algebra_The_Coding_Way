```python
import numpy as np
from numpy.linalg import eig

def eigen_values_vectors(M):
    # Calculate eigenvalues and eigenvectors
    eigenvalues, eigenvectors = eig(M)

    # Print eigenvalues and eigenvectors
    print("Eigenvalues:\n", eigenvalues)
    print("\nEigenvectors:\n", eigenvectors)

    # Concatenate eigenvectors into a matrix
    eig_matrix = np.column_stack(eigenvectors)

    # Extract the real part of eigenvalues
    real_parts = np.real(eigenvalues)

    # Find the index where the real part is closest to 1
    index_of_1 = np.argmin(np.abs(real_parts - 1))

    return np.real(eig_matrix[:, index_of_1])

def scale_vect(V):
    return V / np.sum(V)

def calculate_pagerank(matrix):
    # Calculate the PageRank vector
    pagerank = eigen_values_vectors(matrix)

    # Scale the PageRank vector to make it a probability distribution
    scaled_pagerank = scale_vect(pagerank)

    # Ensure that PageRank values are non-negative
    scaled_pagerank = np.abs(scaled_pagerank)

    return scaled_pagerank

# Define the transition probability matrix
A = np.array([[0, 0, 1, 0.5],
              [1/3, 0, 0, 0],
              [1/3, 0.5, 0, 0.5],
              [1/3, 0.5, 0, 0]])

# Print eigenvalues and eigenvectors
print("Eigenvalues and Eigenvectors:")
pagerank = eigen_values_vectors(A)
print()

Eigenvalues and Eigenvectors:
Eigenvalues:
 [ 1.        +0.j         -0.36062333+0.41097555j -0.36062333-0.41097555j
 -0.27875333+0.j        ]

Eigenvectors:
 [[ 0.72101012+0.j         -0.75521571+0.j         -0.75521571-0.j
   0.50648562+0.j        ]
 [ 0.24033671+0.j          0.3036721 +0.34607247j  0.3036721 -0.34607247j
  -0.60565568+0.j        ]
 [ 0.54075759+0.j          0.09315321-0.2746779j   0.09315321+0.2746779j
  -0.38153917+0.j        ]
 [ 0.36050506+0.j          0.3583904 -0.07139457j  0.3583904 +0.07139457j
   0.48070923+0.j        ]]
# Display the PageRank values for each page
pages = ['Page 1', 'Page 2', 'Page 3', 'Page 4']
print("Importance of each page (scaled PageRank):")
pagerank_normalized = calculate_pagerank(A)
for page, rank in zip(pages, pagerank_normalized):
    print(f"{page}: {rank:.4f}")

Importance of each page (scaled PageRank):
Eigenvalues:
 [ 1.        +0.j         -0.36062333+0.41097555j -0.36062333-0.41097555j
 -0.27875333+0.j        ]

Eigenvectors:
 [[ 0.72101012+0.j         -0.75521571+0.j         -0.75521571-0.j
   0.50648562+0.j        ]
 [ 0.24033671+0.j          0.3036721 +0.34607247j  0.3036721 -0.34607247j
  -0.60565568+0.j        ]
 [ 0.54075759+0.j          0.09315321-0.2746779j   0.09315321+0.2746779j
  -0.38153917+0.j        ]
 [ 0.36050506+0.j          0.3583904 -0.07139457j  0.3583904 +0.07139457j
   0.48070923+0.j        ]]
Page 1: 2.5483
Page 2: 2.6692
Page 3: 2.6692
Page 4: 1.7901


```
