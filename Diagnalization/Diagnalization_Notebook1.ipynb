```python
import numpy as np
import matplotlib.pyplot as plt

# Define the diagonal matrix D
D = np.array([[3, 0],
              [0, 6]])

# Define basis vectors
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# Transform basis vectors using D
e1_transformed = np.dot(D, e1)
e2_transformed = np.dot(D, e2)

# Plot basis vectors before and after transformation
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='b', label='e1')
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='r', label='e2')
plt.quiver(0, 0, e1_transformed[0], e1_transformed[1], angles='xy', scale_units='xy', scale=1, color='g', label='D[e1]')
plt.quiver(0, 0, e2_transformed[0], e2_transformed[1], angles='xy', scale_units='xy', scale=1, color='m', label='D[e2]')
plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Basis vectors before and after transformation')
plt.show()


```

```python
# Transform vector x = [1, 1] using D
x = np.array([1, 1])
x_transformed = np.dot(D, x)

# Plot vector x before and after transformation
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='b', label='x')
plt.quiver(0, 0, x_transformed[0], x_transformed[1], angles='xy', scale_units='xy', scale=1, color='g', label='D[x]')
plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Vector x before and after transformation')
plt.show()
```
