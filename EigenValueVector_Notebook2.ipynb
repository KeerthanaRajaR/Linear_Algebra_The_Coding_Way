```python
Understanding the problem:
          1.Find the Eigenvalues and Eigenvectors of the given matrix.
              a)If we get distinct eigen values,we have to prove that the eigenvectors that corresponds to that eigenvalues are independent vectors.For that ,we have to use Reduced Row Echolon Form (RREF),In that if the number of pivot columns matches the number of distinct eigenvalues, it suggests that the eigenvectors corresponding to distinct eigenvalues are independent.
              b)Otherwise linearly dependent
Step 1:
      Define Matrix A We start by defining the matrix A as a numpy array.

Step 2:
      Calculate Eigenvalues and Eigenvectors Using numpy's np.linalg.eig() function, we calculate the eigenvalues and eigenvectors of matrix A.

Step 3:
      Print Eigenvalues and Eigenvectors We print the calculated eigenvalues and eigenvectors to examine the results.

Step 4:
      Check Independence of Eigenvectors We stack the eigenvectors obtained as columns to form a matrix. Then, we calculate the rank of this matrix using np.linalg.matrix_rank().

Step 5:
      Count Number of Distinct Eigenvalues We count the number of distinct eigenvalues using np.unique() function.

Step 6:
      Print Results We print the number of distinct eigenvalues and the rank of the matrix formed by eigenvectors.

Step 7:
      Verify Independence of Eigenvectors.Finally, we compare the rank of the matrix formed by eigenvectors with the number of distinct eigenvalues. If they are equal, we conclude that the eigenvectors corresponding to distinct eigenvalues are independent; otherwise, we conclude that they are not independent.
```

```python
import numpy as np

# Define the matrix A
A = np.array([[2, 1, 0], [1, 3, -1], [0, -1, 4]])

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Print eigenvalues and eigenvectors
print("Eigenvalues:")
print(eigenvalues)
print("\nEigenvectors:")
print(eigenvectors)

```

```python
# Function to perform row reduction operations
def row_reduce(matrix):
    num_rows, num_cols = matrix.shape
    # Iterate over each row
    for pivot_row in range(num_rows):
        # Find the pivot element
        pivot_element = matrix[pivot_row, pivot_row]
        # Make the pivot element 1
        matrix[pivot_row] /= pivot_element
        # Eliminate non-zero elements below the pivot
        for row in range(pivot_row + 1, num_rows):
            factor = matrix[row, pivot_row]
            matrix[row] -= factor * matrix[pivot_row]

# Perform row reduction
row_reduce(eigenvectors)

# Count the number of non-zero rows
rank = np.count_nonzero(np.sum(np.abs(eigenvectors), axis=1))

# Number of distinct eigenvalues
num_distinct_eigenvalues = len(np.unique(eigenvalues))

# Print results
print("\nNumber of distinct eigenvalues:", num_distinct_eigenvalues)
print("Rank of the matrix formed by eigenvectors:", rank)

# Verify independence of eigenvectors
if rank == num_distinct_eigenvalues:
    print("\nThe eigenvectors corresponding to distinct eigenvalues are independent.")
else:
    print("\nThe eigenvectors corresponding to distinct eigenvalues are not independent.")
```
